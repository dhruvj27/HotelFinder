```python
bengaluru_content = pd.read_csv("/kaggle/input/content-based-data/bengaluru_content_dataset.csv")
delhi_content = pd.read_csv("/kaggle/input/content-based-data/delhi_content_dataset.csv")
hyderabad_content = pd.read_csv("/kaggle/input/content-based-data/hyderabad_content_dataset.csv")
jaipur_content = pd.read_csv("/kaggle/input/content-based-data/jaipur_content_dataset.csv")
pune_content = pd.read_csv("/kaggle/input/content-based-data/pune_content_dataset.csv")
mumbai_content = pd.read_csv("/kaggle/input/content-based-data/mumbai_content_dataset.csv")
chennai_content = pd.read_csv("/kaggle/input/content-based-data/chennai_content_dataset.csv")
kolkata_content = pd.read_csv("/kaggle/input/content-based-data/kolkata_content_dataset.csv")
ahmedabad_content = pd.read_csv("/kaggle/input/content-based-data/ahmedabad_content_dataset.csv")
lucknow_content = pd.read_csv("/kaggle/input/content-based-data/lucknow_content_dataset.csv")
```

```python
final_content_df = pd.concat([bengaluru_content, delhi_content, hyderabad_content, jaipur_content, pune_content, mumbai_content, chennai_content, kolkata_content, ahmedabad_content, lucknow_content], axis=0, ignore_index=True)
final_content_df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
vectorizer = TfidfVectorizer()
feature_matrix = vectorizer.fit_transform(final_content_df["description"])
similarity = cosine_similarity(feature_matrix)
```

```python
def recommend(hotel_name, top_n=5):
    if hotel_name not in final_content_df['display_name'].values:
        return f"❌ Hotel '{hotel_name}' not found in dataset."

    idx = final_content_df[final_content_df['display_name'] == hotel_name].index[0]
    scores = list(enumerate(similarity[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)

    recommendations = final_content_df.iloc[[i[0] for i in scores[1:top_n+1]]][['hotel_id','display_name', 'rating']]
    return recommendations
```

```python
hotel_name = input("Enter a hotel name: ")

print("\nRecommended Hotels:\n")
print(recommend(hotel_name))
```
